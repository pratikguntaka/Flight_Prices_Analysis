## Outlier detection
for numerical values/feature. Identifying outlier with Distinction plot, Box plot, Histogram.  In the graphs/plots, we can see positive or negative outliers based on which way the graphs are skewed. 
IQR - Interquartile Range. One way to deal with an oulier is to replace the outlier value with Median value

```python
def plot(df, col):
    fig, (ax1, ax2, ax3) = plt.subplots(3,1)
    sns.distplot(df[col], ax=ax1)
    sns.boxplot(df[col], ax=ax2)
    sns.distplot(df[col], ax=ax3 , kde=False)
```

```python
plot(train_copy , 'Price')
```

```python
q1 = train_copy['Price'].quantile(0.25)
q3 = train_copy['Price'].quantile(0.75)

iqr = q3-q1

maximum = q3 + 1.5*iqr
minimum = q1 - 1.5*iqr
```
